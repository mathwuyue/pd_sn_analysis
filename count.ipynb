```python
import pandas as pd
df1 = pd.read_csv('final_data.csv')
df1
```

```python
data1 = df1[df1['region'].notna()]
data1
```

```python
import datetime
data1['publish_time_time'] = pd.to_datetime(data1['publish_time'], format='%Y-%m-%dT%H:%M:%S+08:00')
data1['year'] = data1['publish_time_time'].dt.strftime('%Y')
data1['year'].value_counts()
```

```python
data1 = data1.loc[data1['year'] == '2022']
group1 = data1.groupby(["year","region"])
count1 = group1.size().reset_index(name='counts')
count1
```

```python
count1.to_csv('count_year_region_new.csv')
```

```python
data1['month-day'] = data1['publish_time_time'].dt.strftime('%m-%d')
```

```python
group2 = data1.groupby(['month-day', 'region'])
count2 = group2.size().reset_index(name = 'counts')
count2
```

```python
count2.to_csv('count_day_region.csv')
```
