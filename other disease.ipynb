```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import warnings
warnings.filterwarnings('ignore')
df1 = pd.read_csv('first_classify_data.csv')
df2 = pd.read_csv('other_disease.csv')
```

```python
counts1 = df1.groupby('pre_labels').size().reset_index(name='counts')
counts1['key'] = '帕金森'
# 0 refers advertisements, 1 refers the opposite
```

```python
counts2 = df2.groupby(['key','pre_labels']).size().reset_index(name='counts').drop([0,1]).reset_index(drop=True)
counts2
```

```python
stats = pd.DataFrame(columns=['disease','ad-ratio','non-ad-ratio'])
for name in set(counts2['key']):
    tmp = counts2.loc[counts2['key'] == name,['pre_labels','counts']].reset_index(drop=True)
    ratio0 = tmp.iloc[0,1]/tmp['counts'].sum()
    ratio1 = tmp.iloc[1,1]/tmp['counts'].sum()
    stats.loc[len(stats)] = [name, ratio0,ratio1]
ratio0 = counts1.iloc[0,1]/ counts1['counts'].sum()
ratio1 =  counts1.iloc[1,1]/ counts1['counts'].sum()
stats.loc[len(stats)] = ['帕金森', ratio0,ratio1]
stats
```

```python
def summary_statistics(df):
    df_summary = df.describe().T
    df_summary['std'] = df.var()
    return df_summary
summary_statistics(stats.drop([8]))
```

```python
from scipy.stats import chi2_contingency
import numpy as np
stats = pd.DataFrame(columns=['疾病','p-value'])
for name in set(counts2['key']):
    tmp = counts2.loc[counts2['key'] == name].reset_index(drop=True)
    tmp = pd.concat([tmp, counts1],axis=1)
    tmp = tmp[['counts','counts-1']].T
    print(name)
    print(tmp)
    print('\n')
    _, p_value, _, _ = chi2_contingency(tmp)
    stats.loc[len(stats)] = [name, np.round(p_value,5)]
stats
```
